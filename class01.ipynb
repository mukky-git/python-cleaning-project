# üçï 100 Days of Python ‚Äî Conditional Statements Practice

## Day 12: Adding Toppings

### üß© Challenge

You‚Äôre continuing your pizza order program.  
Now you‚Äôll **add toppings and calculate the final bill** using **nested conditionals**.

---

### üîπ Instructions

1. Ask the user to choose a pizza size:  
   - `S` for Small  
   - `M` for Medium  
   - `L` for Large  

   - Prices:
     - Small Pizza = **$15**
     - Medium Pizza = **$20**
     - Large Pizza = **$25**

2. Ask the user if they want **pepperoni**:  
   - If **yes**:
     - Add **$2** for a small pizza  
     - Add **$3** for medium or large pizza  

3. Ask the user if they want **extra cheese**:  
   - If **yes**, add **$1** regardless of size.

4. Finally, display their **total bill** in a friendly message.

---

### üß† Example Interaction

`````python
print("Welcome to Python Pizza Deliveries")

size = input("What size pizza do you want? S, M, or L: ")
add_pepperoni = input("Do you want pepperoni? Y or N: ")
extra_cheese = input("Do you want extra cheese? Y or N: ")
print(size)

bill = 0

if size == "S":
    bill = 15
elif size == "M":
    bill = 20
elif size == "L":
    bill = 25
else:
    print("Invalid size chosen")

if add_pepperoni == "Y":
    if size == "S":
        bill += 2
    else:
        bill += 3

if extra_cheese == "Y":
    bill += 1

print(f"Your final bill is: ${bill}")
print(size)
print(f"the pizza you ordered is {size} and you are paying ${bill}")

`````
