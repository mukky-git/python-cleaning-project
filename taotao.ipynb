OOP stands for Object-Oriented Programming. It's a way of designing programs using objects and classes.

- Object: A real-world entity (e.g., a dog, a car, a student).

- Class: A blueprint/template for creating objects.

### Why Use OOP?
1. To model real-world things more naturally.

2. Reusability of code through inheritance.

3. Easier maintenance.

4. Encapsulation (hiding details and exposing only what‚Äôs necessary).

5. Polymorphism (using one interface for multiple forms).

## Basic Terminologies in OOP (Python)

| **Term**      | **Meaning**                                  |
|---------------|----------------------------------------------|
| `Class`       | Blueprint or template                        |
| `Object`      | Instance of a class                          |
| `Method`      | Function inside a class                      |
| `Attribute`   | Variable associated with a class/object      |
| `Constructor` | Method that runs when an object is created: `__init__` |


- Classes are a way by which, in object-oriented programming, we can create our own type of data and give them names.
- A class is like a mold for a type of data ‚Äì where we can invent our own data type and give them a name.
- We can modify our code as follows to implement our own class called Student:

Any time you create a class and you utilize that blueprint to create something, you create what is called an ‚Äúobject‚Äù or an ‚Äúinstance‚Äù. In the case of our code, student is an object.

- The constructor in Python is the `__init__()` method.
- It automatically runs when you create a new object from a class.

**Its purpose**:

To initialize (set up) the object‚Äôs data when the object is created.

````python
class Student:
    def __init__(self, name_of_student, house_of_student):
        self.name = name_of_student
        self.house = house_of_student
````

`self` is a reference to the current object.

It lets you:

- Access and modify the object's data and methods inside the class.

- Differentiate between instance variables (self.x) and local variables (name).

Here:

- name is a parameter passed in

- self.name means ‚Äúthis specific student‚Äôs name‚Äù

So self.name = name means:

> "Set the object‚Äôs name to whatever name was passed in."

````python
student = Student("Harry Porter", "Griffindor")
student.house
````
