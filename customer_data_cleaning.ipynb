```python
import pandas as pd

df = pd.read_csv("customer_sales_raw.csv")
df.head(18)
```

```python
## this give me information about the data am working on 
df.info()
```

```python
# with my df.info()
# i was able to know that i have 8 email missing 
```

```python
##count how many missing exist in each column
df.isnull().sum()
```

```python
# we have been able to drop our duplicated rows from 310 to 300
df = df.drop_duplicates()
df.shape
```

```python
df.info()
```

```python
#we had 8 missing mails and now they would all be filled up with unknown@example.com
df['Email'] = df['Email'].fillna('unknown@example.com')
```

```python
#now we dont have any missing value 
df.info()
```

```python
# Remove spaces
df['Email'] = df['Email'].str.strip()

# Replace double @@ with single @
df['Email'] = df['Email'].str.replace('@@', '@')

# Optional: convert all emails to lowercase
df['Email'] = df['Email'].str.lower()
```

```python
# Remove quotes and extra symbols, then title case
df['Name'] = df['Name'].str.replace('"','').str.replace('!',' ').str.replace('@',' ').str.strip().str.title()
```

```python
df['Country'] = df['Country'].str.strip().str.lower().replace({
    'nigeria':'Nigeria',
    'ghana':'Ghana',
    'kenya':'Kenya'
})
```

ðŸ”¹ Explanation

pd.to_datetime(df['Date']) â†’ converts strings to datetime objects.

dayfirst=True â†’ interprets dates like 26/06/21 as 26 June 2021.

errors='coerce' â†’ if a date is invalid (like 2021-28-09), it becomes NaT (Not a Time).

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
```

```python
df[df['Date'].isna()]
```

```python
## i dropped 6 dates column that where not having date and now we are left with 294
df = df.dropna(subset=['Date'])
df.shape
```

```python
## we are going to be removing negative signs right now from the amount row 
df['Amount'] = df['Amount'].abs()
```

```python
df.shape
```

```python
df.to_csv('customer_sales_cleaned.csv', index=False)
```
